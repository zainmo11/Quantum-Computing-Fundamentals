# Bell states

```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector
from IPython.display import display, Markdown
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
import numpy as np
import matplotlib.pyplot as plt
from qiskit.visualization import plot_histogram
```

```python

# Bell state labels and their corresponding operations (I, X, Z, XZ)
bell_preparations = {
    'Ψ+ (|00⟩+|11⟩)': [],
    'Ψ- (|00⟩−|11⟩)': ['x_1'],
    'Φ+ (|01⟩+|10⟩)': ['x_0'],
    'Φ- (|01⟩−|10⟩)': ['x_0', 'x_1']
}
```

```python
for label, ops in bell_preparations.items():
    qr = QuantumRegister(2)
    qc = QuantumCircuit(qr)

    # Step 2: Apply operations to turn it into other Bell states
    for op in ops:
      if op == 'x_1':
          qc.x(qr[1])
      if op == 'x_0':
          qc.x(qr[0])

    # Step 1: Prepare initial Bell pair Φ+
    qc.h(qr[1])
    qc.cx(qr[1], qr[0])


    # Show Bell state
    display(Markdown(f"### Bell State: {label}"))
    bell_state = Statevector.from_instruction(qc)
    display(bell_state.draw('latex'))
    display(qc.draw('mpl', reverse_bits=True))
```

# Superdense Coding Circuit

```python
# all 2 digits pairs
all_pairs = ['00','01','10','11']
# Store results from all pairs
all_counts = {}
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from IPython.display import display, Markdown

# Correct Bell state operations: targeting both q[0] and q[1] appropriately
bell_preparations = {
    'Ψ+ (|00⟩+|11⟩)': [],
    'Ψ- (|00⟩−|11⟩)': ['x_1'],
    'Φ+ (|01⟩+|10⟩)': ['x_0'],
    'Φ- (|01⟩−|10⟩)': ['x_0', 'x_1']
}

# Classical message pairs for superdense coding
all_pairs = ['00', '01', '10', '11']
all_counts = {}

# Loop over Bell state preparations
for bell_label, bell_ops in bell_preparations.items():
    display(Markdown(f"## Initial Bell State: {bell_label}"))

    # Loop over all classical message pairs
    for pair in all_pairs:
        qr = QuantumRegister(2, 'q')
        cr = ClassicalRegister(2, 'c')
        qc = QuantumCircuit(qr, cr)

        # Apply Bell state modifiers as per image
        for op in bell_ops:
            if op == 'x_1':
                qc.x(qr[1])
            if op == 'x_0':
                qc.x(qr[0])

        # Prepare the initial entangled state: always from |00⟩
        qc.h(qr[1])
        qc.cx(qr[1], qr[0])

        qc.barrier(label="Entanglement Created")

        # Alice encodes classical bits
        if bell_label ==  'Ψ+ (|00⟩+|11⟩)':
          if pair[0] == '1':
              qc.z(qr[1])
          if pair[1] == '1':
              qc.x(qr[1])
        elif bell_label == 'Φ- (|01⟩−|10⟩)':
          if pair[0] == '0':
              qc.z(qr[1])
          if pair[1] == '0':
              qc.x(qr[1])
        elif bell_label == 'Φ+ (|01⟩+|10⟩)':
          if pair[0] == '1':
              qc.z(qr[1])
          if pair[1] == '0':
              qc.x(qr[1])
        elif bell_label == 'Ψ- (|00⟩−|11⟩)':
          if pair[0] == '0':
              qc.z(qr[1])
          if pair[1] == '1':
              qc.x(qr[1])
        qc.barrier(label="Alice Encodes Bits")

        # Bob decodes
        qc.cx(qr[1], qr[0])
        qc.h(qr[1])
        qc.barrier(label="Bob Decodes")

        # Measurement
        qc.measure(qr, cr)

        # Simulate
        sim = AerSimulator()
        job = sim.run(qc, shots=100)
        result = job.result()
        counts = result.get_counts()
        all_counts[(bell_label, pair)] = counts

        display(Markdown(f"### Message: `{pair}`"))
        display(qc.draw('mpl',reverse_bits=True))
        print(f"Counts for Bell State '{bell_label}' with pair {pair}: {counts}")
```

# Quantum Teleportation Protocol

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import random_statevector, partial_trace , DensityMatrix, Statevector
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_bloch_multivector
import numpy as np
```

```python
# create quantum state
state = random_statevector(2)
print("Initial state")
display(state.draw('latex', prefix="|\\psi\\rangle ="))
display(plot_bloch_multivector(state))
```

```python
# create quantum and classical registers
qr = QuantumRegister(3,'q')
cr = ClassicalRegister(2,'c')
qc = QuantumCircuit(qr, cr)
print(qc)
```

```python
# initialize q2 with the state to be teleported
qc.initialize(state, qr[2])
qc.barrier()
```

```python
# create Bell pair between Alice --> q1 and bob --> q0
qc.h(qr[1])
qc.cx(qr[1], qr[0])
qc.barrier(label="$|\\psi_{\\text{pre}}\\rangle$")
```

```python
# Alice performs Bell-state measurment
qc.cx(qr[2], qr[1])
qc.h(qr[2])
qc.barrier(label="$|\\psi_{\\text{mid}}\\rangle$")
```

```python
# measurement
qc.measure(qr[2], cr[1])
qc.measure(qr[1], cr[0])
```

```python
# conditinal operations
with qc.if_test((cr[0], 1)):
    qc.x(qr[0])
with qc.if_test((cr[1], 1)):
    qc.z(qr[0])
```

```python
# save final statevector for analysis
qc.save_statevector(label="$|\\psi_{\\text{final}}\\rangle$")
```

```python
display(qc.draw('mpl', reverse_bits=True, cregbundle=False))
```

```python
# simulate
simulator = AerSimulator()
job = simulator.run(qc, shots=1)
result = job.result()
counts = result.get_counts()
print("Measurement results:")
print(counts)
```

```python
final_state = Statevector(result.data(0)["$|\\psi_{\\text{final}}\\rangle$"])
print("final Global Stateevector pre-measurment:")
display(final_state.draw('latex', prefix="|\\psi_{pre}\\rangle = "))
display(plot_bloch_multivector(final_state))
```

```python
final_state
```

```python
bob_qubit = partial_trace(final_state,[1,2])
print("Bob's qubit after measurement:")
display(bob_qubit.draw('latex', prefix="|\\psi_{bob}\\rangle = "))
display(plot_bloch_multivector(bob_qubit))
```

```python
from qiskit.quantum_info import state_fidelity
from qiskit.quantum_info import Statevector, partial_trace
ordinal_dm =DensityMatrix(state)
fidelity_value = state_fidelity(bob_qubit, ordinal_dm)
print(f"State fidelity: {fidelity_value:.14f}")

reduced_dm = [final_state.data[0] , final_state.data[1]]
reduced_dm = Statevector(np.array(reduced_dm))

compare_state = reduced_dm ^ state

display(plot_bloch_multivector(compare_state, title="Bob's qubit Before / After measurement"))
```


> Qubit teleported correctly




# Quantum Random Number Generator

```python
from qiskit import QuantumCircuit
import matplotlib.pyplot as plt
simulator = AerSimulator()

random_numbers = []
```

```python
# Generate 100 quantum random bits
for _ in range(100):
    # Create a quantum circuit with 1 qubit and 1 classical bit
    qc = QuantumCircuit(1, 1)
    qc.h(0)
    qc.measure(0, 0)

    # Execute the circuit
    job = simulator.run(qc, shots=1)
    result = job.result()
    counts = result.get_counts(qc)
    bit = int(list(counts.keys())[0])
    random_numbers.append(bit)

print(random_numbers)
```

```python
# Plot the distribution of 0s and 1s
plt.hist(random_numbers, bins=[-0.5, 0.5, 1.5], rwidth=0.8, align='mid', color='skyblue', edgecolor='black')
plt.xticks([0, 1])
plt.xlabel('Quantum Bit Value')
plt.ylabel('Frequency')
plt.title('Distribution of 100 Quantum Random Bits')
plt.grid(True)
plt.show()
```



> The values 1 and 0 are equally likely, resulting in a uniform distribution.

## Full Implementation

```python
# Set number of qubits
n = 16
qr = QuantumRegister(n, 'qr')
cr = ClassicalRegister(n, 'cr')
```

```python
# Create quantum circuit for Alice
alice = QuantumCircuit(qr, cr, name='Alice')

# Generate random number as Alice's key and convert it to binary
alice_key = np.random.randint(0, 2**n)
alice_key_binary = format(alice_key, f'0{n}b')

# Encode the key as Alice's qubits
for i, bit in enumerate(alice_key_binary):
    if bit == '1':
        alice.x(qr[i])

# Randomly apply Hadamard gate to half of the qubits
alice_table = []
for i in range(n):
    if np.random.random() < 0.5:
        alice.h(qr[i])
        alice_table.append('X')
    else:
        alice_table.append('Z')

display(alice.draw('mpl',reverse_bits=True))
```

```python
# Create quantum circuit for Bob
bob = QuantumCircuit(qr, cr, name='Bob')

# Bob randomly chooses a basis for measurement
bob_table = []
for i in range(n):
    if np.random.random() < 0.5:
        bob.h(qr[i])
        bob_table.append('X')
    else:
        bob_table.append('Z')

display(bob.draw('mpl',reverse_bits=True))
```

```python
# Measure qubits in Bob's circuit
bob.measure(qr, cr)

# Combine Alice's and Bob's circuits
full_circuit = QuantumCircuit(qr, cr)
full_circuit.compose(alice, inplace=True)
full_circuit.compose(bob, inplace=True)

```

```python
# Execute on a local simulator
simulator = AerSimulator()
result = simulator.run(full_circuit, shots=1).result()


```

```python
# Extract Bob's key
counts = result.get_counts()
bob_key = list(counts.keys())[0]
bob_key = bob_key[::-1]  # Reverse to match the qubits' order
```

```python
# Compare basis and determine which qubits to keep
keep = []
discard = []
for i, (alice_basis, bob_basis) in enumerate(zip(alice_table, bob_table)):
    if alice_basis == bob_basis:
        keep.append(i)
    else:
        discard.append(i)

# Extract the sifted keys
new_alice_key = [alice_key_binary[i] for i in keep]
new_bob_key = [bob_key[i] for i in keep]
```

```python
# Print the results
print(f'Percentage of qubits to be discarded: {len(discard) / n}')
print(f'New Alice\'s key: {new_alice_key}')
print(f'New Bob\'s key: {new_bob_key}')

# Check if the keys match
if new_alice_key == new_bob_key:
    print("Key exchange successful!")
else:
    print("Key exchange tampered! Check for eavesdropper.")
```
